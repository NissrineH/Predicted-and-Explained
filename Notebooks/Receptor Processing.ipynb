```python
aa_alphabet = {'A': 1, 'C': 2, 'D': 3, 'E': 4, 'F': 5, 'G': 6, 'H': 7, 'I': 8, 'K': 9, 'L': 10, 'M': 11, 'N': 12, 'P': 13, 'Q': 14, 'R': 15, 'S': 16, 'T': 17, 'V': 18, 'W': 19, 'Y': 20}

def encode_and_pad_sequences(seq_list):
    encoded_seqs = []
    max_length = max(len(seq) for seq in seq_list)
    for seq in seq_list:
        encoded_seq = [aa_alphabet.get(aa, 0) for aa in seq]
        padding = [0] * (max_length - len(seq))
        encoded_seq += padding
        encoded_seqs.append(encoded_seq)
    return encoded_seqs

seq = [sequences['6IIU'], sequences['2ZV2'], sequences['6D6T'], sequences['5EK0'], sequences['4F8H'], sequences['1T7R']]
encoded_seqs = encode_and_pad_sequences(seq)

seq_df = pd.DataFrame(list(sequences.items()), columns=['pdb_id',  'sequence'])
seq_df['encoded_seq'] = encoded_seqs
```
